```python
def get_company_info(wd):
    company = pd.DataFrame(
        columns=['name', 'summary', 'description', 'followerCount', 'industry', 'companySize',
                 'headquarters','type','specialties'])
    
    # visiting the about page
    current_url = wd.current_url
    own_company = current_url.find('applaudo-studios')
    if own_company != -1 :
        wd.get('https://www.linkedin.com/company/applaudo-studios/mycompany/?viewAsMember=true')
    else:
        about = current_url + 'about'

    wd.get(about)
    
    top_container = wd.find_element_by_class_name('block.mt2')
    name = top_container.find_element_by_xpath('//h1').text
    top_cont_info = top_container.find_elements_by_class_name('org-top-card-summary-info-list__info-item')
    for info in top_cont_info:
        followers_div = info.text.find('follower')
        if followers_div != -1:
            followers = info.text.split(' ')[0]
            break
        else:
            followers = ''
            
    try:
        summary = top_container.find_element_by_class_name('org-top-card-summary__tagline.t-16.t-black').text
    except:
        summary = ''
        
    overview_elmn = wd.find_element_by_class_name('artdeco-card.p4.mb3')
    try:
        description_cn = 'break-words.white-space-pre-wrap.mb5.t-14.t-black--light.t-normal'
        description = overview_elmn.find_element_by_class_name(description_cn).text
    except:
        description = ''

    details = overview_elmn.find_elements_by_class_name('org-page-details__definition-term.t-14.t-black.t-bold')
    details_txt = overview_elmn.find_elements_by_class_name('org-page-details__definition-text.t-14.t-black--light.t-normal')
    
    industry, headquarters, ctype, specialties = '', '', '', ''

    for i in range(len(details)):
        if details[i].text == 'Industry':
            industry = details_txt[i].text
        elif details[i].text == 'Headquarters':
            headquarters = details_txt[i-1].text
        elif details[i].text == 'Type':
            ctype = details_txt[i-1].text
        elif details[i].text == 'Specialties':
            specialties = details_txt[i-1].text
    
    try:
        company_size_cn = 'org-about-company-module__company-size-definition-text.t-14.t-black--light.mb1.fl'
        company_size = wd.find_element_by_class_name(company_size_cn).text
    except:
        company_size = ''
    
    try:
        employees_li_cn = 'org-page-details__employees-on-linkedin-count.t-14.t-black--light.mb5'
        employees_linkedin = wd.find_element_by_class_name(employees_li_cn).text.split(' ')[0]
    except:
        employees_linkedin = ''
        
    print(name)
    print(industry)
    print(company_size)
    print(headquarters)
    print(ctype)
    print(specialties)
    print(employees_linkedin)
        
    
    company = company.append({'name': name, 'summary': summary, 'description': description,
                              'followerCount': followers, 'industry': industry, 'companySize': company_size,
                              'headquarters': headquarters, 'type': ctype, 'specialties': specialties},
                             ignore_index=True)
    
#     print(company)
    return company
    
```
